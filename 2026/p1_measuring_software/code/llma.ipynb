```python
import os, re, time, threading, gc
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from itertools import combinations
from scipy import stats

from llama_cpp import Llama
from pynvml import (
    nvmlInit, nvmlShutdown, nvmlDeviceGetHandleByIndex,
    nvmlDeviceGetPowerUsage, nvmlDeviceGetUtilizationRates,
    nvmlDeviceGetMemoryInfo, nvmlDeviceGetTemperature, NVML_TEMPERATURE_GPU
)

N_CTX = 1024
MAX_TOKENS = 256
PROMPT = "Explain quantization tradeoffs in LLMs in 6 bullet points."
SAMPLE_INTERVAL_S = 0.1

N_RUNS = 30
N_WARMUP = 3
ALPHA = 0.05
MICRO_REST_S = 60


TEMPERATURE = 0.0
TOP_P = 1.0
SEED = 42

def qlabel(name: str) -> str:
    m = re.search(r"(Q\d+[_A-Z]*)", str(name))
    return m.group(1) if m else str(name)

def _collect_gpu_samples(stop_event, samples, device_index=0, interval_s=0.05):
    nvmlInit()
    h = nvmlDeviceGetHandleByIndex(device_index)
    t0 = time.time()
    try:
        while not stop_event.is_set():
            now = time.time()
            power_mw = nvmlDeviceGetPowerUsage(h)
            util = nvmlDeviceGetUtilizationRates(h).gpu
            mem = nvmlDeviceGetMemoryInfo(h)
            temp_c = nvmlDeviceGetTemperature(h, NVML_TEMPERATURE_GPU)
            samples.append({
                "t": now - t0,
                "power_w": power_mw / 1000.0,
                "util_gpu": util,
                "mem_used_mib": mem.used / (1024**2),
                "temp_c": float(temp_c),
            })
            time.sleep(interval_s)
    finally:
        nvmlShutdown()

def energy_joules_from_samples(df_samples: pd.DataFrame) -> float:
    if df_samples is None or df_samples.empty or df_samples.shape[0] < 2:
        return 0.0
    t = df_samples["t"].to_numpy()
    p = df_samples["power_w"].to_numpy()
    return float(((p[1:] + p[:-1]) / 2.0 * (t[1:] - t[:-1])).sum())

def load_model(model_path: str, seed: int = 42) -> Llama:
    return Llama(
        model_path=model_path,
        n_ctx=N_CTX,
        n_gpu_layers=-1,
        verbose=False,
        seed=seed,
    )

def _one_infer_unmeasured(llm: Llama):
    _ = llm.create_chat_completion(
        messages=[
            {"role": "system", "content": "You are a concise, correct assistant."},
            {"role": "user", "content": PROMPT},
        ],
        temperature=TEMPERATURE,
        top_p=TOP_P,
        max_tokens=MAX_TOKENS,
    )

def bench_inference_only(llm: Llama) -> dict:
    stop_event = threading.Event()
    samples = []
    th = threading.Thread(
        target=_collect_gpu_samples,
        args=(stop_event, samples, 0, SAMPLE_INTERVAL_S),
        daemon=True,
    )
    th.start()

    t0 = time.time()
    out = llm.create_chat_completion(
        messages=[
            {"role": "system", "content": "You are a concise, correct assistant."},
            {"role": "user", "content": PROMPT},
        ],
        temperature=TEMPERATURE,
        top_p=TOP_P,
        max_tokens=MAX_TOKENS,
    )
    t1 = time.time()

    stop_event.set()
    th.join(timeout=2.0)

    text = out["choices"][0]["message"]["content"]
    usage = out.get("usage", {})
    gen_tokens = usage.get("completion_tokens", None)
    if gen_tokens is None:
        gen_tokens = len(llm.tokenize(text.encode("utf-8")))

    seconds = t1 - t0
    tok_per_sec = gen_tokens / seconds if seconds > 0 else float("inf")

    df_samples = pd.DataFrame(samples)
    energy_j = energy_joules_from_samples(df_samples)
    j_per_tok = energy_j / gen_tokens if gen_tokens else float("inf")

    mean_temp_c = float(df_samples["temp_c"].mean()) if not df_samples.empty else float("nan")
    max_temp_c  = float(df_samples["temp_c"].max())  if not df_samples.empty else float("nan")

    return {
        "seconds": seconds,
        "gen_tokens": gen_tokens,
        "tok_per_sec": tok_per_sec,
        "energy_j": energy_j,
        "j_per_token": j_per_tok,
        "mean_temp_c": mean_temp_c,
        "max_temp_c": max_temp_c,
        "gpu_trace": df_samples,
        "output_text": text,
    }

def _interp_trace_to_grid(df_trace: pd.DataFrame, t_grid: np.ndarray, col: str) -> np.ndarray:
    if df_trace is None or df_trace.empty:
        return np.full_like(t_grid, np.nan, dtype=float)
    t = df_trace["t"].to_numpy()
    y = df_trace[col].to_numpy().astype(float)
    order = np.argsort(t)
    t, y = t[order], y[order]
    y_grid = np.full_like(t_grid, np.nan, dtype=float)
    mask = (t_grid >= t[0]) & (t_grid <= t[-1])
    if mask.any():
        y_grid[mask] = np.interp(t_grid[mask], t, y)
    return y_grid

def average_traces(traces: list, col: str, n_points: int = 600):
    end_times = [float(tr["t"].max()) for tr in traces if tr is not None and not tr.empty]
    if not end_times:
        return np.array([0.0, 1.0]), np.array([np.nan, np.nan]), np.array([np.nan, np.nan])
    t_max = max(end_times)
    t_grid = np.linspace(0.0, t_max, n_points)
    ys = np.vstack([_interp_trace_to_grid(tr, t_grid, col) for tr in traces])
    return t_grid, np.nanmean(ys, axis=0), np.nanstd(ys, axis=0)

def cohens_d(x, y):
    x = np.asarray(x, dtype=float); y = np.asarray(y, dtype=float)
    x = x[~np.isnan(x)]; y = y[~np.isnan(y)]
    nx, ny = len(x), len(y)
    if nx < 2 or ny < 2:
        return np.nan
    sx, sy = np.std(x, ddof=1), np.std(y, ddof=1)
    sp = np.sqrt(((nx - 1)*sx*sx + (ny - 1)*sy*sy) / (nx + ny - 2))
    return (np.mean(x) - np.mean(y)) / sp if sp > 0 else np.nan

def holm_correction(pvals):
    pvals = np.asarray(pvals, dtype=float)
    m = len(pvals)
    order = np.argsort(pvals)
    adj = np.empty(m, dtype=float)
    for k, idx in enumerate(order):
        adj[idx] = min((m - k) * pvals[idx], 1.0)
    adj_sorted = adj[order]
    for i in range(1, m):
        adj_sorted[i] = max(adj_sorted[i], adj_sorted[i-1])
    adj[order] = adj_sorted
    return adj

def pairwise_stats_quant(df: pd.DataFrame, metric: str, quant_order: list, alpha=0.05) -> pd.DataFrame:
    pairs = list(combinations(quant_order, 2))
    raw_p, rows = [], []
    for a, b in pairs:
        xa = df.loc[df["quant"] == a, metric].to_numpy(dtype=float)
        xb = df.loc[df["quant"] == b, metric].to_numpy(dtype=float)
        xa = xa[~np.isnan(xa)]; xb = xb[~np.isnan(xb)]
        t, p = stats.ttest_ind(xa, xb, equal_var=False, nan_policy="omit")
        d = cohens_d(xa, xb)
        raw_p.append(float(p))
        rows.append({"A": a, "B": b, "welch_t": float(t), "p_raw": float(p), "cohens_d": float(d)})
    p_holm = holm_correction(raw_p)
    for i in range(len(rows)):
        rows[i]["p_holm"] = float(p_holm[i])
        rows[i]["significant"] = bool(p_holm[i] < alpha)
    return pd.DataFrame(rows).sort_values("p_holm")

model_order = list(paths.keys())
quant_order = [qlabel(m) for m in model_order]
_seen = set()
quant_order = [q for q in quant_order if not (q in _seen or _seen.add(q))]

traces_by_model = {m: [] for m in model_order}
all_runs = []
load_times = {m: [] for m in model_order}
outputs_by_model = {m: [] for m in model_order}

for r in range(N_RUNS + N_WARMUP):
    is_warmup = (r < N_WARMUP)
    label = "WARMUP" if is_warmup else "MEASURED"
    shown_idx = (r + 1) if is_warmup else (r - N_WARMUP + 1)
    shown_total = N_WARMUP if is_warmup else N_RUNS

    print(f"\n{'='*50}\n{label} Round {shown_idx}/{shown_total}\n{'='*50}")

    for model_name in model_order:
        model_path = paths[model_name]

        print(f"  Loading {model_name}...")
        t0 = time.time()
        llm = load_model(model_path, seed=SEED)
        t1 = time.time()
        load_times[model_name].append(t1 - t0)
        print(f"    Load time: {t1 - t0:.2f}s")

        if is_warmup:
            print("    Running warmup inference (not recorded)...")
            _one_infer_unmeasured(llm)
        else:
            print("    Running measured inference...")
            res = bench_inference_only(llm)

            all_runs.append({
                "model": model_name,
                "round": shown_idx,
                "seconds": res["seconds"],
                "gen_tokens": res["gen_tokens"],
                "tok_per_sec": res["tok_per_sec"],
                "energy_j": res["energy_j"],
                "j_per_token": res["j_per_token"],
                "mean_temp_c": res["mean_temp_c"],
                "max_temp_c": res["max_temp_c"],
            })
            traces_by_model[model_name].append(res["gpu_trace"])
            outputs_by_model[model_name].append(res["output_text"])
            print(f"    Done: {res['gen_tokens']} tokens, {res['energy_j']:.1f}J, {res['tok_per_sec']:.1f} tok/s")

        del llm
        gc.collect()

        if MICRO_REST_S > 0:
            time.sleep(MICRO_REST_S)

df_runs = pd.DataFrame(all_runs)
df_runs["quant"] = df_runs["model"].map(qlabel)
df_runs["quant"] = pd.Categorical(df_runs["quant"], categories=quant_order, ordered=True)


print("\n" + "="*50)
print("OUTPUT DETERMINISM VERIFICATION")
print("="*50)

all_det = True
for m in model_order:
    outs = outputs_by_model[m]
    ok = len(set(outs)) == 1
    chars = len(outs[0]) if outs else 0
    print(f"  {qlabel(m)}: {'✓ DETERMINISTIC' if ok else '✗ VARIES'} ({chars} chars)")
    if not ok:
        all_det = False

print("\n" + ("✓ All outputs deterministic." if all_det else "✗ WARNING: Some outputs vary."))

df_load = pd.DataFrame([
    {"model": qlabel(m), "mean_load_s": np.mean(load_times[m]), "std_load_s": np.std(load_times[m])}
    for m in model_order
])
print("\n" + "="*50)
print("MODEL LOADING TIMES (excluded from energy)")
print("="*50)
print(df_load.to_string(index=False))

metrics = ["seconds", "gen_tokens", "tok_per_sec", "energy_j", "j_per_token", "mean_temp_c", "max_temp_c"]
df_mean = df_runs.groupby("quant")[metrics].mean().reindex(quant_order).reset_index()

print("\n" + "="*50)
print(f"INFERENCE METRICS (mean of {N_RUNS} runs; warmup {N_WARMUP} excluded)")
print("="*50)
print(df_mean.to_string(index=False))


print("\n" + "="*50)
print("Shapiro–Wilk + Kruskal–Wallis + Welch Pairwise (Holm) + Cohen's d")
print("="*50)

stats_metrics = [
    ("seconds", "Inference Time (s)"),
    ("tok_per_sec", "Tokens per Second"),
    ("energy_j", "GPU Energy per Inference (J)"),
    ("j_per_token", "Energy per Token (J/token)"),
]

for met, title in stats_metrics:
    print(f"\n--- {title} ---")

    groups = [df_runs.loc[df_runs["quant"] == q, met].dropna().to_numpy(dtype=float) for q in quant_order]

    print("Shapiro–Wilk (per quant):")
    for q, g in zip(quant_order, groups):
        if len(g) < 3:
            print(f"  {q}: n={len(g)}  (skipped; need >=3)")
            continue
        W, p_sh = stats.shapiro(g)
        print(f"  {q}: n={len(g)}  W={W:.4f}  p={p_sh:.4g}  normal={p_sh > ALPHA}")

    H, p_kw = stats.kruskal(*groups)
    print(f"Kruskal–Wallis: H = {H:.6f}, p = {p_kw:.6e}, significant = {p_kw < ALPHA}")

    pw = pairwise_stats_quant(df_runs, met, quant_order, alpha=ALPHA)
    print(pw[["A","B","welch_t","p_raw","cohens_d","p_holm","significant"]].to_string(index=False))


fig, axes = plt.subplots(2, 3, figsize=(15, 10))
box_metrics = [
    ("seconds", "Inference Time (s)"),
    ("tok_per_sec", "Tokens / Second"),
    ("energy_j", "Energy (J)"),
    ("j_per_token", "J / Token"),
    ("mean_temp_c", "Mean GPU Temp (°C)"),
    ("max_temp_c", "Max GPU Temp (°C)"),
]

for ax, (met, title) in zip(axes.flat, box_metrics):
    data = [df_runs.loc[df_runs["quant"] == q, met].dropna().to_numpy() for q in quant_order]

    ax.violinplot(
        data,
        showmeans=False,
        showmedians=False,
        showextrema=False
    )

    ax.boxplot(
        data,
        widths=0.18,
        patch_artist=True,
        showfliers=False
    )

    ax.set_title(title)
    ax.set_xlabel("Quantization")
    ax.set_xticks(range(1, len(quant_order) + 1))
    ax.set_xticklabels(quant_order)

plt.suptitle(
    f"LLM Quantization Benchmark: Inference Only\n({N_RUNS} runs; warmup {N_WARMUP} excluded; loading excluded)",
    fontsize=14
)
plt.tight_layout()
plt.show()


plt.figure(figsize=(18, 10))
plt.suptitle(f"Llama.cpp Quantization Benchmark ({N_RUNS} runs avg): GPU traces + metrics", fontsize=16)

ax1 = plt.subplot(2, 1, 1)
for m in model_order:
    traces = traces_by_model[m]
    t, mu, sd = average_traces(traces, col="power_w", n_points=600)
    ax1.plot(t, mu, label=qlabel(m))
    ax1.fill_between(t, mu - sd, mu + sd, alpha=0.15)
ax1.set_title(f"GPU Power vs Time (mean ± std over {N_RUNS} runs)")
ax1.set_xlabel("Time (s)")
ax1.set_ylabel("Power (W)")
ax1.legend(loc="upper right", ncol=6)

ax2 = plt.subplot(2, 1, 2)
for m in model_order:
    traces = traces_by_model[m]
    t, mu, sd = average_traces(traces, col="util_gpu", n_points=600)
    ax2.plot(t, mu, label=qlabel(m))
    ax2.fill_between(t, mu - sd, mu + sd, alpha=0.15)
ax2.set_title(f"GPU Utilization vs Time (mean ± std over {N_RUNS} runs)")
ax2.set_xlabel("Time (s)")
ax2.set_ylabel("GPU Util (%)")
ax2.legend(loc="lower right", ncol=6)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()


summary = df_mean[["quant", "seconds", "tok_per_sec", "energy_j", "j_per_token", "mean_temp_c"]].copy()
summary = summary.round({"seconds": 3, "tok_per_sec": 2, "energy_j": 2, "j_per_token": 4, "mean_temp_c": 1})

print("\n" + "="*50)
print("FINAL SUMMARY (inference only, excludes loading)")
print("="*50)
print(summary.to_string(index=False))
```
