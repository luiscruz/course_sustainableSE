```python
from dotenv import load_dotenv
load_dotenv()
import os
USERNAMME = os.getenv("USERNAMME")
PASSWORD = os.getenv("PASSWORD")
EMAIL = os.getenv("EMAIL")
print(f"Username: {USERNAMME}")
print(f"Password: {PASSWORD}")
print(f"Email: {EMAIL}")
```

```python
import requests
register_url = 'https://api.watttime.org/register'
params = {'username': USERNAMME,
         'password': PASSWORD,
         'email': EMAIL}
rsp = requests.post(register_url, json=params)
print(rsp.text)
```

```python
from requests.auth import HTTPBasicAuth
login_url = 'https://api.watttime.org/login'
rsp = requests.get(login_url, auth=HTTPBasicAuth(USERNAMME, PASSWORD))
TOKEN = rsp.json()['token']
print(rsp.json())
```

```python
url = "https://api.watttime.org/v3/region-from-loc"
headers = {"Authorization": f"Bearer {TOKEN}"}
params = {"latitude": "52.132", "longitude": "5.291", "signal_type": "co2_moer"}
response = requests.get(url, headers=headers, params=params)
response.raise_for_status()
print(response.json())
```
